```python
auto = pd.read_csv('C:\\Users\\Systems\\Automobile_data.csv')
auto.head()
```

```python
auto.info()
```

```python
auto.replace('?',np.nan,inplace = True)
```

```python
auto['price'] = pd.to_numeric(auto['price'], errors='coerce')
auto['horsepower'] = pd.to_numeric(auto['horsepower'], errors='coerce')
auto['peak-rpm'] = pd.to_numeric(auto['peak-rpm'], errors='coerce')
```

```python
auto.price = auto.price.apply(float)
auto.horsepower = auto.horsepower.apply(float)
auto['peak-rpm'] = auto['peak-rpm'].apply(float)
```

```python
auto.nunique()
```

```python
auto.duplicated().sum() #To check duplicated value amd sum
```

```python
auto.isnull().sum()
```

```python
print("Fuel Types          : ",auto['fuel-type'].unique())
print("Body Style          : ",auto['body-style'].unique())
print("Wheel Drives        : ",auto['drive-wheels'].unique())
print("Engine Location     : ",auto['engine-location'].unique())
print("Aspiration          : ",auto['aspiration'].unique())
print("Number of Doors     : ",auto['num-of-doors'].unique())
print("Number of Cylinders : ",auto['num-of-cylinders'].unique())
print("Engine Types        : ",auto['engine-type'].unique())
print("Fuel System         : ",auto['fuel-system'].unique())
print("Manufacture         : ",auto['make'].unique())
```
