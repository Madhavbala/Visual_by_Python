```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

### Plot 
  - creates a plot for numerical data

```python
x = np.arange(10)
y = x *2 
print(x,y)
plt.figure() # figure used to dscribe the size on the graph
plt.plot(x,y)
```

```python
plt.plot(x,y,'g-o')
```

```python
a = np.linspace(1,10,1000)
b = np.sin(a)
plt.figure()
plt.plot(a,b,'#f3ff05')#search color picker in google to get the code
plt.title("Title")
plt.xlabel("X label")
plt.ylabel("Y label")
```

```python
a = np.linspace(1,10,1000)
b = np.sin(a)
plt.figure()
plt.plot(a,b,'#f3ff05')#search color picker in google to get the code
plt.title("Title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.xticks(range(0,11,2))
plt.yticks(range(0,11,2))
plt.show()
```

```python
a = np.linspace(1,10,1000)
b = np.sin(a)
plt.figure()
plt.plot(a,b,'#00d9ff')#search color picker in google to get the code
plt.plot(a,b+2,'#00d9ff')#search color picker in google to get the code
plt.title("Title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.xticks(range(0,11,2))
plt.yticks(range(0,11))
plt.show()
```

```python
a = np.linspace(1,10,1000)
b = np.sin(a)
plt.figure()
plt.xticks(range(0,11,2))
plt.yticks(range(0,11))
plt.title("Title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.plot(a,b,'#00d9ff' , label = "X Power")#search color picker in google to get the code
plt.plot(a,b+2,'r--',label = "Y Power")#search color picker in google to get the code
plt.legend()
plt.show()
```

### Scatter

```python
x = np.arange(60)
y = x + np.random.randint(1,10,60)
print(x,y)
plt.scatter(x,y,label = "Random Data")
plt.legend()
```

### Bar

```python
students = ['Madhav','Madhavan','Aakash','Mathu']
marks = [10,20,10,20]
plt.bar(students,marks)
```

### Pie

```python
x = [1,20,5,40,50]
values = ['bike','car','plane','train','bus']
plt.pie(x,labels = values)
plt.plot()
```

### Barplot

```python
salaries = np.random.randint(5000,50000,500)
plt.boxplot(salaries)
plt.show()
```

### Violinplot

```python
salaries = np.random.randint(123123,3242993,454)
plt.violinplot(salaries)
plt.show
```

### Histogram

```python
x = np.arange(1,15)
plt.hist(x,bins=9)
plt.show()
```
